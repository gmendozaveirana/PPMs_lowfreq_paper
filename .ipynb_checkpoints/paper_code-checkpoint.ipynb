```python
"""
################################## PERMITTIVITY MODELS EVALUATION ##################################
"""
""" 
Defining parameters after laboratory analysis and calculations 
"""
samples = ['EH26', 'A44', 'valthe5', 'EH23', 'P17', 'Dren8', 'E44', 'D348', 'hul586', 'valthe11']

#   chequear la curva A44 si con otro procesado safa porque el m y alpha dan medio raros

s = 0

for sample in samples:
    s += 1
    
    ############################## Definning some common parameters for each soil sample ######################
    
    print("Sample:", sample)
    etot =    np.sum((dt.loc[:, sample+"_p"]-np.mean(dt.loc[:, sample+"_p"]))**2)   # Total sum of squares
    sc =      ds.at[s, "Sand_content"]               # SAND content (63 - 2000 * e-6 meters) (%)
    silt =    ds.at[s, "Silt_content"]               # SILT content (%)
    cc  =     ds.at[s, "Clay_content"]               # CLAY content (0  - 2    * e-6 meters) %[m3/m3]
    bd =      ds.at[s, "Bulk_density_reached"]       # Bulk density reached during the laboratory sample preparation the same as measured in field conditions.
    ps =      ds.at[s, "Solid_phase_permittivity"]   # Solid phase permittivity calculated using CRIM model, after experiment.
    org =     ds.at[s, "Humus"]                      # Humus (organic matter) content 
    bw =      ds.at[s, "Bound_water"]                # Bound water, after experiment [m3/m3]  
    CEC =     ds.at[s, "CEC"]                       # Cation Exchange Capacity [meq/100g]
    Initial_perm = ds.at[s, "Initial_perm"]
    Initial_water = ds.at[s, "Initial_water"]
    """ 
    Calculing parameters using the previous defined parameters
    """
    pdn =     ptf.schjonnpd(cc, org)                  # partile density of layer(s) (g/cm)
    por = 1 - (bd/pdn)                   

    print("Particle density of "+sample+" after ptf.schjonnpd:", str("{:.2f}".format(pdn)))

    """ 
    Evaluating fitting permittivity models using the previous defined parameters
    """
    
    ################################  Fitting models evaluation ######################################
    
    Roth90_RMSE, LindeP_RMSE, EndresRedman_RMSE, FengSen_RMSE, Wund13_RMSE, Sen81_RMSE, Dobson_RMSE = [], [], [], [], [], [], []

    """ Roth's model"""
    for i in range(len(alpha_)):
        
        Roth90_RMSE.append(np.sqrt(np.mean((ptf.roth1990( dt.loc[:, sample+"_w"], bd, pdn, ap, ps, dt.loc[:, sample+"_wpc"], alpha_[i])/e_0-dt.loc[:, sample+"_p"])**2)))    

    Roth90_RMSE_min = min(Roth90_RMSE)
    alpha_opt = alpha_[Roth90_RMSE.index(Roth90_RMSE_min)]
    print('alpha optim:', alpha_opt)
    
    Roth90_opt = ptf.roth1990( dt.loc[:, sample+"_w"], bd, pdn, ap, ps, dt.loc[:, sample+"_wpc"], alpha_opt)/e_0
    FitsP.at["Roth90_opt", "RMSE_"+sample] = np.sqrt(np.mean((Roth90_opt-dt.loc[:, sample+"_p"])**2))
    FitsP.at["Roth90_opt", "R2_"+sample] =   1 - (np.sum((Roth90_opt-dt.loc[:, sample+"_p"])**2)/etot)
    ds.at[s, "Alpha"] = alpha_opt
    
    """ Embedding schemes """
    for i in range(len(L_)):
        EndresRedman_RMSE.append(np.sqrt(np.mean((ptf.endresredman1996(dt.loc[:, sample+"_w"], bd, pdn, L_[i], ap, ps, dt.loc[:, sample+"_wpc"])/e_0-dt.loc[:, sample+"_p"])**2)))
        FengSen_RMSE.append(np.sqrt(np.mean((ptf.fengsen1985(          dt.loc[:, sample+"_w"], bd, pdn, L_[i], ap, ps, dt.loc[:, sample+"_wpc"])/e_0-dt.loc[:, sample+"_p"])**2)))
        Wund13_RMSE.append(np.sqrt(np.mean((ptf.wunderlich2013perm(    dt.loc[:, sample+"_w"], Initial_perm, Initial_water,                L_[i]            , dt.loc[:, sample+"_wpc"])/e_0-dt.loc[:, sample+"_p"])**2)))
        Sen81_RMSE.append(np.sqrt(np.mean((ptf.sen1981(                dt.loc[:, sample+"_w"], bd, pdn, L_[i], ap, ps, dt.loc[:, sample+"_wpc"])/e_0-dt.loc[:, sample+"_p"])**2)))

    EndresRedman_RMSE_min = min(EndresRedman_RMSE)
    FengSen_RMSE_min = min(FengSen_RMSE)
    Wund13_RMSE_min = min(Wund13_RMSE)
    Sen81_RMSE_min = min(Sen81_RMSE)

    L_opt_EndresRedman = L_[EndresRedman_RMSE.index(EndresRedman_RMSE_min)]
    L_opt_FengSen      = L_[FengSen_RMSE.index(FengSen_RMSE_min)]
    L_opt_Wund13       = L_[Wund13_RMSE.index(Wund13_RMSE_min)]
    L_opt_Sen81        = L_[Sen81_RMSE.index(Sen81_RMSE_min)]
    
    EndresRedman_opt = ptf.endresredman1996(  dt.loc[:, sample+"_w"], bd, pdn, L_opt_EndresRedman, ap, ps, dt.loc[:, sample+"_wpc"])/e_0
    FengSen_opt = ptf.fengsen1985(       dt.loc[:, sample+"_w"], bd, pdn, L_opt_FengSen, ap, ps, dt.loc[:, sample+"_wpc"])/e_0
    Wund13_opt = ptf.wunderlich2013perm(dt.loc[:, sample+"_w"], Initial_perm, Initial_water,                 L_opt_Wund13, dt.loc[:, sample+"_wpc"])/e_0
    Sen81_opt = ptf.sen1981(           dt.loc[:, sample+"_w"], bd, pdn, L_opt_Sen81, ap, ps, dt.loc[:, sample+"_wpc"]    )/e_0
    
    EndresRedman = ptf.endresredman1996(  dt.loc[:, sample+"_w"], bd, pdn, 0.3, ap, ps, dt.loc[:, sample+"_wpc"])/e_0
    FengSen = ptf.fengsen1985(       dt.loc[:, sample+"_w"], bd, pdn, 0.3, ap, ps, dt.loc[:, sample+"_wpc"])/e_0
    Wund13 = ptf.wunderlich2013perm(dt.loc[:, sample+"_w"], 4, 0.05,          0.01, dt.loc[:, sample+"_wpc"])/e_0
    Sen81 = ptf.sen1981(           dt.loc[:, sample+"_w"], bd, pdn, 0.3, ap, ps, dt.loc[:, sample+"_wpc"]    )/e_0

    FitsP.at["EndresRedman_opt", "RMSE_"+sample] = np.sqrt(np.mean((EndresRedman_opt-dt.loc[:, sample+"_p"])**2))
    FitsP.at["FengSen_opt", "RMSE_"+sample]      = np.sqrt(np.mean((FengSen_opt-dt.loc[:, sample+"_p"])**2))
    FitsP.at["Wund13_opt", "RMSE_"+sample]       = np.sqrt(np.mean((Wund13_opt-dt.loc[:, sample+"_p"])**2))
    FitsP.at["Sen81_opt", "RMSE_"+sample]        = np.sqrt(np.mean((Sen81_opt-dt.loc[:, sample+"_p"])**2))
        
    FitsP.at["EndresRedman_opt", "R2_"+sample]   = 1 - (np.sum((EndresRedman_opt-dt.loc[:, sample+"_p"])**2)/etot)
    FitsP.at["FengSen_opt", "R2_"+sample]        = 1 - (np.sum((EndresRedman_opt-dt.loc[:, sample+"_p"])**2)/etot)
    FitsP.at["Wund13_opt", "R2_"+sample]         = 1 - (np.sum((Wund13_opt-dt.loc[:, sample+"_p"])**2)/etot)                
    FitsP.at["Sen81_opt", "R2_"+sample]          = 1 - (np.sum((Sen81_opt-dt.loc[:, sample+"_p"])**2)/etot)

    print("The optimum L EndresRedman is:", L_opt_EndresRedman)
    print("The optimum L FengSen is:", L_opt_FengSen )
    print("The optimum L Wund13 is:", L_opt_Wund13)
    print("The optimum L Sen81 is:", L_opt_Sen81)
    ds.at[s, "L_EndresRedman"] = L_opt_EndresRedman
    ds.at[s, "L_FengSen"] = L_opt_FengSen
    ds.at[s, "L_Wund13"] = L_opt_Wund13
    ds.at[s, "L_Sen81"] = L_opt_Sen81

    """ Linde's model"""
    for i in range(len(m_)):
        LindeP_RMSE.append(np.sqrt(np.mean((ptf.lindeperm(dt.loc[:, sample+"_w"], bd, pdn, ap, ps, dt.loc[:, sample+"_wpc"], m_[i], m_[i])/e_0-dt.loc[:, sample+"_p"])**2)))

    LindeP_RMSE_min = min(LindeP_RMSE)
    m_opt = m_[LindeP_RMSE.index(LindeP_RMSE_min)]
    
    LindeP_opt = ptf.lindeperm(dt.loc[:, sample+"_w"], bd, pdn, ap, ps, dt.loc[:, sample+"_wpc"], m_opt, m_opt)/e_0
    LindeP = ptf.lindeperm(dt.loc[:, sample+"_w"], bd, pdn, ap, ps, dt.loc[:, sample+"_wpc"], 1.5, 2)/e_0
    FitsP.at["LindeP_opt", "RMSE_"+sample] = np.sqrt(np.mean((LindeP_opt-dt.loc[:, sample+"_p"])**2))
    FitsP.at["LindeP_opt", "R2_"+sample]   = 1 - (np.sum((LindeP_opt-dt.loc[:, sample+"_p"])**2)/etot)
    print("The optimum m is:", m_opt)
    ds.at[s, "m_linde"] = m_opt

    FitsP

    """ Dobson's model"""
    for i in range(len(bwp_)):
        
        Dobson_RMSE.append(np.sqrt(np.mean((ptf.dobson85(dt.loc[:, sample+"_w"], bw, bd, pdn, ap, ps, dt.loc[:, sample+"_wpc"], bwp_[i])/e_0 -dt.loc[:, sample+"_p"])**2)))

    Dobson_RMSE_min = min(Dobson_RMSE)
    bwp_opt = bwp_[Dobson_RMSE.index(Dobson_RMSE_min)]
    
    Dobson_opt = ptf.dobson85(dt.loc[:, sample+"_w"], bw, bd, pdn, ap, ps, dt.loc[:, sample+"_wpc"], bwp_opt)/e_0
    FitsP.at["Dobson_opt", "RMSE_"+sample] = np.sqrt(np.mean(( Dobson_opt - dt.loc[:, sample+"_p"])**2))
    FitsP.at["Dobson_opt", "R2_"+sample] =   1 - (np.sum((Dobson_opt-dt.loc[:, sample+"_p"])**2)/etot)
    print("The optimum bwp is:", bwp_opt)
    ds.at[s, "bwp_dobson"] = bwp_opt
    
    
    ####################################### Non-fitting models evaluation ################################
    
    
    Hydra_pred, Topp_pred, Jacandschj1993B_pred, Nadler_pred, Logsdonperm_pred, Steelman_pred, Jacandschj1993A_pred = [], [], [], [], [], [], []
    CRIM_pred, Peplinski95_pred, Malicki_pred= [], [], [], 
    FengSen_Mod_pred, Roth90_Mod_pred, LindeP_Mod_pred, Sen81_Mod_pred, Dobson_mod_pred, EndresRedman_mod_pred, Wund13_mod_pred = [], [], [], [], [], [], []
    
    Hydra_pred = ptf.Hydraprobe(      dt.loc[:, sample+"_w"]  )/e_0
    Topp_pred =  ptf.topp1980(dt.loc[:, sample+"_w"].all())/e_0
    Jacandschj1993B_pred =  ptf.jacandschj1993B(      dt.loc[:, sample+"_w"].any(), bd, cc, org  )/e_0
    Nadler_pred = ptf.nadler1991(      dt.loc[:, sample+"_w"].any()  )/e_0
    #Logsdonperm_pred = ptf.logsdonperm(      dt.loc[:, sample+"_w"].any()  )/e_0
    Steelman_pred = ptf.steelman(      dt.loc[:, sample+"_w"].any()  )/e_0
    Jacandschj1993A_pred = ptf.jacandschj1993A(      dt.loc[:, sample+"_w"].any()  )/e_0
    
    FitsP.at["Hydra", "RMSE_"+sample]  =          np.sqrt(np.mean((Hydra_pred - dt.loc[:, sample+"_p"])**2))
    FitsP.at["Topp",  "RMSE_"+sample] =           np.sqrt(np.mean(( Topp_pred - dt.loc[:, sample+"_p"])**2))
    FitsP.at["Jacandschj1993B", "RMSE_"+sample] = np.sqrt(np.mean(( Jacandschj1993B_pred - dt.loc[:, sample+"_p"])**2))
    FitsP.at["Nadler", "RMSE_"+sample] =          np.sqrt(np.mean(( Nadler_pred - dt.loc[:, sample+"_p"])**2))
    #FitsP.at["Logsdonperm", "RMSE_"+sample] =     np.sqrt(np.mean(( Logsdonperm_pred - dt.loc[:, sample+"_p"])**2))
    FitsP.at["Steelman", "RMSE_"+sample] =        np.sqrt(np.mean(( Steelman_pred - dt.loc[:, sample+"_p"])**2))
    FitsP.at["Jacandschj1993A", "RMSE_"+sample] = np.sqrt(np.mean(( Jacandschj1993A_pred - dt.loc[:, sample+"_p"])**2))
    
    CRIM_pred = ptf.CRIM(            dt.loc[:, sample+"_w"], bd, pdn,            ap, ps, dt.loc[:, sample+"_wpc"])/e_0
    Peplinski95_pred = ptf.peplinski(       dt.loc[:, sample+"_w"], bd, pdn, cc, sc,    ps, fq, dt.loc[:, sample+"_wpc"], p_inf)/e_0
    Malicki_pred = ptf.malicki(         dt.loc[:, sample+"_w"], bd                                )/e_0
    
    FitsP.at["CRIM", "RMSE_"+sample] =            np.sqrt(np.mean((CRIM_pred - dt.loc[:, sample+"_p"])**2))
    FitsP.at["Peplinski95", "RMSE_"+sample] =     np.sqrt(np.mean((Peplinski95_pred - dt.loc[:, sample+"_p"])**2))
    FitsP.at["Malicki", "RMSE_"+sample] =         np.sqrt(np.mean((Malicki_pred - dt.loc[:, sample+"_p"])**2))
   
    Roth90_IOP = ptf.roth1990_IOP(dt.loc[:, sample+"_w"], bd, pdn,  ap, ps, dt.loc[:, sample+"_wpc"], CEC)/e_0
    LindeP_IOP = ptf.lindep_IOP(dt.loc[:, sample+"_w"], bd, pdn,   ap, ps, dt.loc[:, sample+"_wpc"], CEC)/e_0
    Sen81_IOP = ptf.sen81_IOP(dt.loc[:, sample+"_w"], bd, pdn,    ap, ps, dt.loc[:, sample+"_wpc"], CEC)/e_0
    #Dobson_IOP 
    endresredman_IOP = ptf.endresredman_IOP(dt.loc[:, sample+"_w"], bd, pdn, ap, ps, dt.loc[:, sample+"_wpc"], CEC)/e_0
    fengsen_IOP = ptf.fengsen_IOP(dt.loc[:, sample+"_w"], bd, pdn, ap, ps, dt.loc[:, sample+"_wpc"], CEC)/e_0
    Wund13_IOP = ptf.wund13_IOP(dt.loc[:, sample+"_w"], 4, 0.05, dt.loc[:, sample+"_wpc"], CEC)/e_0
    
    FitsP.at["FengSen_IOP", "RMSE_"+sample] =      np.sqrt(np.mean((fengsen_IOP - dt.loc[:, sample+"_p"])**2))
    FitsP.at["Roth90_IOP", "RMSE_"+sample] =       np.sqrt(np.mean((Roth90_IOP - dt.loc[:, sample+"_p"])**2))
    FitsP.at["LindeP_IOP", "RMSE_"+sample] =       np.sqrt(np.mean((LindeP_IOP - dt.loc[:, sample+"_p"])**2))
    FitsP.at["Sen81_IOP", "RMSE_"+sample] =        np.sqrt(np.mean((Sen81_IOP - dt.loc[:, sample+"_p"])**2))
    #FitsP.at["Dobson_mod", "RMSE_"+sample] =       np.sqrt(np.mean((Dobson_mod_pred-dt.loc[:, sample+"_p"])**2))
    FitsP.at["EndresRedman_IOP", "RMSE_"+sample] = np.sqrt(np.mean((endresredman_IOP - dt.loc[:, sample+"_p"])**2))
    FitsP.at["Wund13_IOP", "RMSE_"+sample] =       np.sqrt(np.mean((Wund13_IOP - dt.loc[:, sample+"_p"])**2))


    FitsP.at["Hydra", "R2_"+sample]  =          1 - (np.sum((Hydra_pred-dt.loc[:, sample+"_p"])**2)/etot)                                               
    FitsP.at["Topp",  "R2_"+sample] =           1 - (np.sum((Topp_pred-dt.loc[:, sample+"_p"])**2)/etot)
    FitsP.at["Jacandschj1993B", "R2_"+sample] = 1 - (np.sum((Jacandschj1993B_pred-dt.loc[:, sample+"_p"])**2)/etot)
    FitsP.at["Nadler", "R2_"+sample] =          1 - (np.sum((Nadler_pred-dt.loc[:, sample+"_p"])**2)/etot)
    #FitsP.at["Logsdonperm", "R2_"+sample] =     1 - (np.sum((LindeP_pred-dt.loc[:, sample+"_p"])**2)/etot)
    FitsP.at["Steelman", "R2_"+sample] =        1 - (np.sum((Steelman_pred-dt.loc[:, sample+"_p"])**2)/etot)
    FitsP.at["Jacandschj1993A", "R2_"+sample] = 1 - (np.sum((Jacandschj1993A_pred-dt.loc[:, sample+"_p"])**2)/etot)

    FitsP.at["CRIM", "R2_"+sample] =            1 - (np.sum((CRIM_pred-dt.loc[:, sample+"_p"])**2)/etot)
    FitsP.at["Peplinski95", "R2_"+sample] =     1 - (np.sum((Peplinski95_pred-dt.loc[:, sample+"_p"])**2)/etot)
    FitsP.at["Malicki", "R2_"+sample] =         1 - (np.sum((Malicki_pred-dt.loc[:, sample+"_p"])**2)/etot)

    FitsP.at["FengSen_IOP", "R2_"+sample] =      1 - (np.sum((fengsen_IOP-dt.loc[:, sample+"_p"])**2)/etot)
    FitsP.at["Roth90_IOP", "R2_"+sample] =       1 - (np.sum((Roth90_IOP - dt.loc[:, sample+"_p"])**2)/etot)
    FitsP.at["LindeP_IOP", "R2_"+sample] =       1 - (np.sum((LindeP_IOP - dt.loc[:, sample+"_p"])**2)/etot)
    FitsP.at["Sen81_IOP", "R2_"+sample] =        1 - (np.sum((Sen81_IOP  - dt.loc[:, sample+"_p"])**2)/etot)
    #FitsP.at["Dobson_IOP", "R2_"+sample] =       1 - (np.sum((LindeP_IOP-dt.loc[:, sample+"_p"])**2)/etot)
    FitsP.at["EndresRedman_IOP", "R2_"+sample] = 1 - (np.sum((endresredman_IOP-dt.loc[:, sample+"_p"])**2)/etot)
    FitsP.at["Wund13_IOP", "R2_"+sample] =       1 - (np.sum((Wund13_IOP-dt.loc[:, sample+"_p"])**2)/etot)
    
    FitsP.at["FengSen", "R2_"+sample] =      1 - (np.sum((FengSen-dt.loc[:, sample+"_p"])**2)/etot)
    FitsP.at["LindeP", "R2_"+sample] =       1 - (np.sum((LindeP - dt.loc[:, sample+"_p"])**2)/etot)
    FitsP.at["Sen81", "R2_"+sample] =        1 - (np.sum((Sen81 - dt.loc[:, sample+"_p"])**2)/etot)
    #FitsP.at["Dobson", "R2_"+sample] =       1 - (np.sum((LindeP_IOP-dt.loc[:, sample+"_p"])**2)/etot)
    FitsP.at["EndresRedman", "R2_"+sample] = 1 - (np.sum((EndresRedman-dt.loc[:, sample+"_p"])**2)/etot)
    FitsP.at["Wund13", "R2_"+sample] =       1 - (np.sum((Wund13-dt.loc[:, sample+"_p"])**2)/etot)
    
    FitsP.at["FengSen", "RMSE_"+sample] =      np.sqrt(np.mean((FengSen - dt.loc[:, sample+"_p"])**2))
    FitsP.at["LindeP", "RMSE_"+sample] =       np.sqrt(np.mean((LindeP - dt.loc[:, sample+"_p"])**2))
    FitsP.at["Sen81", "RMSE_"+sample] =        np.sqrt(np.mean((Sen81 - dt.loc[:, sample+"_p"])**2))
    #FitsP.at["Dobson", "RMSE_"+sample] =       np.sqrt(np.mean((Dobson_mod_pred-dt.loc[:, sample+"_p"])**2))
    FitsP.at["EndresRedman", "RMSE_"+sample] = np.sqrt(np.mean((EndresRedman - dt.loc[:, sample+"_p"])**2))
    FitsP.at["Wund13", "RMSE_"+sample] =       np.sqrt(np.mean((Wund13 - dt.loc[:, sample+"_p"])**2))

    
    print(FitsP)
    
print(ds)
```

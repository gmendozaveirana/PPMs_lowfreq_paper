#### This notebook shows the performance of permittivity pedophysical models compiled on ppm.py. The test is performed using laboratory data obtained by us in form of calibration curves for diferent soil types. Some models are fitted to obtain optimum geometrical parameters, and then pedotransfer functions are proposed to calculate them. Novel ppm are tested considering these pedotransfer functions.

```python
""" 
################################ Reading laboratory sample analysis ############################
"""

ds = pd.read_excel("Samples Data.xlsx")
ds                                         
```

```python
# Defining a DataFrame to fill out with model's errors, considering both RMSE and R2 score.
PPM_list = [ "Topp", "Nadler", "Jacandschj1993A", "Jacandschj1993B", "Malicki", "Hydra", "Logsdonperm", "Steelman",    
          "Peplinski95", "crim", "LindeP", 
          "Sen81", "FengSen", "EndresRedman", "Wund13", 
          "Roth90_fitted", "LindeP_fitted", "Sen81_fitted", "FengSen_fitted", "EndresRedman_fitted", "Wund13_fitted", 'Dobson_fitted',
          "Roth90_IOP", "LindeP_IOP", "Sen81_IOP", "feng_sen85_IOP", "endres_redman_IOP", "wund_perm_IOP"]    

FitsP= pd.DataFrame(columns=[ "RMSE_"+ds.Sample_name[1], "R2_"+ds.Sample_name[1], 
                             "RMSE_"+ds.Sample_name[2], "R2_"+ds.Sample_name[2], 
                             "RMSE_"+ds.Sample_name[3], "R2_"+ds.Sample_name[3], 
                             "RMSE_"+ds.Sample_name[4], "R2_"+ds.Sample_name[4], 
                             "RMSE_"+ds.Sample_name[5], "R2_"+ds.Sample_name[5], 
                             "RMSE_"+ds.Sample_name[6], "R2_"+ds.Sample_name[6], 
                             "RMSE_"+ds.Sample_name[7], "R2_"+ds.Sample_name[7],
                             "RMSE_"+ds.Sample_name[8], "R2_"+ds.Sample_name[8], 
                             "RMSE_"+ds.Sample_name[9], "R2_"+ds.Sample_name[9], 
                             "RMSE_"+ds.Sample_name[10], "R2_"+ds.Sample_name[10]   ],
                    
# create a dicctionary in the ppm.py that contains the names of the models here and the function that refers to                
index = PPM_list )
```

```python
"""
################################## PERMITTIVITY PEDOPHYSICAL MODELS EVALUATION ##################################
"""
""" 
Defining a few fix parameters 
"""
ap =       1.3                                              # Air permittivity phase
p_inf =    3.5                                              # Permittivity relative of the soil at infinity frecuency

"""
Defining default synthetic soil physical parameters

"""
vmc_ =           np.arange(0.0025, .5 , (.5 - 0.0025)/100)  # Volumetric water content
alpha_ =         np.arange(0.1,   1.5,  (1.4 )       /50)  # All posible alpha values (geometric parameter)
L_ =             np.arange(-0.2,  0.8,  (1)          /2 )  # All posible depolarization factor (L) values (geometric parameter)
m_ =             np.arange(0.5,   2,    (2 - 0.5)    /50)  # All posible cementation factor (m) values (geometric parameter)
bwp_ =           np.arange(-5000, 2000, 7000         /50)  # All posible bound water values 

samples = ds['Sample_name'].to_list()[1:]                   # List of sample's names

s = 0
for sample in samples:                                      # This iterates over all the ten samples       
    s += 1
    
    ############################## Definning some parameters for each soil sample ######################
    
    print("Sample:", sample)
    sc =      ds.at[s, "Sand_content"]               # SAND content (63 - 2000 * e-6 meters) (%)
    silt =    ds.at[s, "Silt_content"]               # SILT content (%)
    cc  =     ds.at[s, "Clay_content"]               # CLAY content (0  - 2    * e-6 meters) %[m3/m3]
    bd =      ds.at[s, "Bulk_density_reached"]       # Bulk density reached during the laboratory sample preparation the same as measured in field conditions.
    ps =      ds.at[s, "Solid_phase_permittivity"]   # Solid phase permittivity calculated using crim model, after experiment.
    org =     ds.at[s, "Humus"]                      # Humus (organic matter) content 
    bw =      ds.at[s, "Bound_water"]                # Bound water, after experiment [m3/m3]  
    CEC =     ds.at[s, "CEC"]                        # Cation Exchange Capacity [meq/100g]
    
    pdn =     ppm.schjonnpd(cc, org)                 # partile density of layer(s) (g/cm)

    ################################  Fitting models evaluation ######################################
   
    etot =    np.sum((dt.loc[:, sample+"_p"]-np.mean(dt.loc[:, sample+"_p"]))**2)   # Total sum of squares
    Roth90_RMSE, LindeP_RMSE, EndresRedman_RMSE, FengSen_RMSE, Wund13_RMSE, Sen81_RMSE, Dobson_RMSE = [], [], [], [], [], [], [] # Definition of lists

    
    """ Roth's model evaluation """
    for i in range(len(alpha_)):
        
        Roth90_RMSE.append(np.sqrt(np.mean((ppm.roth1990( dt.loc[:, sample+"_w"], bd, pdn, ap, ps, dt.loc[:, sample+"_wpc"], alpha_[i])/epsilon_0-dt.loc[:, sample+"_p"])**2)))    

    Roth90_RMSE_min = min(Roth90_RMSE)
    alpha_fitted = alpha_[Roth90_RMSE.index(Roth90_RMSE_min)]
    
    Roth90_fitted = ppm.roth1990( dt.loc[:, sample+"_w"], bd, pdn, ap, ps, dt.loc[:, sample+"_wpc"], alpha_fitted)/epsilon_0
    FitsP.at["Roth90_fitted", "RMSE_"+sample] = Roth90_RMSE_min
    FitsP.at["Roth90_fitted", "R2_"+sample] =   1 - (np.sum((Roth90_fitted-dt.loc[:, sample+"_p"])**2)/etot)
    ds.at[s, "Alpha"] = alpha_fitted
    
    
    """ Embedding schemes """
    for i in range(len(L_)):
        EndresRedman_RMSE.append(np.sqrt(np.mean((ppm.endres_redman(dt.loc[:, sample+"_w"], bd, pdn, ap, ps, dt.loc[:, sample+"_wpc"], L_[i]) / epsilon_0-dt.loc[:, sample+"_p"])**2)))
        FengSen_RMSE.append(np.sqrt(np.mean((ppm.feng_sen85(          dt.loc[:, sample+"_w"], bd, pdn, ap, ps, dt.loc[:, sample+"_wpc"], L_[i]) / epsilon_0-dt.loc[:, sample+"_p"])**2)))
        Wund13_RMSE.append(np.sqrt(np.mean((ppm.wund_perm(    dt.loc[:, sample+"_w"], dt.loc[:, sample+"_p"].min(), dt.loc[:, sample+"_w"].min(), dt.loc[:, sample+"_wpc"],  L_[i]) / epsilon_0-dt.loc[:, sample+"_p"])**2)))
        Sen81_RMSE.append(np.sqrt(np.mean((ppm.sen1981(                dt.loc[:, sample+"_w"], bd, pdn, ap, ps, dt.loc[:, sample+"_wpc"], L_[i]) / epsilon_0-dt.loc[:, sample+"_p"])**2)))
    
    FitsP.at["EndresRedman_fitted", "RMSE_"+sample] = min(EndresRedman_RMSE)
    FitsP.at["FengSen_fitted", "RMSE_"+sample]      = min(FengSen_RMSE)
    FitsP.at["Wund13_fitted", "RMSE_"+sample]       = min(Wund13_RMSE)
    FitsP.at["Sen81_fitted", "RMSE_"+sample]        = min(Sen81_RMSE)

    L_fitted_EndresRedman = L_[EndresRedman_RMSE.index(min(EndresRedman_RMSE))]
    L_fitted_FengSen      = L_[FengSen_RMSE.index(min(FengSen_RMSE))]
    L_fitted_Wund13       = L_[Wund13_RMSE.index(min(Wund13_RMSE))]
    L_fitted_Sen81        = L_[Sen81_RMSE.index(min(Sen81_RMSE))]
    
    EndresRedman_fitted = ppm.endres_redman(  dt.loc[:, sample+"_w"], bd, pdn, ap, ps, dt.loc[:, sample+"_wpc"], L_fitted_EndresRedman)/epsilon_0
    FengSen_fitted = ppm.feng_sen85(       dt.loc[:, sample+"_w"], bd, pdn, ap, ps, dt.loc[:, sample+"_wpc"], L_fitted_FengSen)/epsilon_0
    Wund13_fitted = ppm.wund_perm(dt.loc[:, sample+"_w"], dt.loc[:, sample+"_p"].min(), dt.loc[:, sample+"_w"].min(), dt.loc[:, sample+"_wpc"], L_fitted_Wund13)/epsilon_0
    Sen81_fitted = ppm.sen1981(           dt.loc[:, sample+"_w"], bd, pdn, ap, ps, dt.loc[:, sample+"_wpc"], L_fitted_Sen81)/epsilon_0
        
    FitsP.at["EndresRedman_fitted", "R2_"+sample]   = 1 - (np.sum((EndresRedman_fitted-dt.loc[:, sample+"_p"])**2)/etot)
    FitsP.at["FengSen_fitted", "R2_"+sample]        = 1 - (np.sum((EndresRedman_fitted-dt.loc[:, sample+"_p"])**2)/etot)
    FitsP.at["Wund13_fitted", "R2_"+sample]         = 1 - (np.sum((Wund13_fitted-dt.loc[:, sample+"_p"])**2)/etot)                
    FitsP.at["Sen81_fitted", "R2_"+sample]          = 1 - (np.sum((Sen81_fitted-dt.loc[:, sample+"_p"])**2)/etot)

    ds.at[s, "L_EndresRedman"] = L_fitted_EndresRedman
    ds.at[s, "L_FengSen"] = L_fitted_FengSen
    ds.at[s, "L_Wund13"] = L_fitted_Wund13
    ds.at[s, "L_Sen81"] = L_fitted_Sen81

    
    """ Linde's model"""
    for i in range(len(m_)):
        LindeP_RMSE.append(np.sqrt(np.mean((ppm.linde_perm(dt.loc[:, sample+"_w"], bd, pdn, ap, ps, dt.loc[:, sample+"_wpc"], m_[i], m_[i])/epsilon_0-dt.loc[:, sample+"_p"])**2)))

    m_fitted = m_[LindeP_RMSE.index(min(LindeP_RMSE))]
    
    LindeP_fitted = ppm.linde_perm(dt.loc[:, sample+"_w"], bd, pdn, ap, ps, dt.loc[:, sample+"_wpc"], m_fitted, m_fitted)/epsilon_0
    LindeP = ppm.linde_perm(dt.loc[:, sample+"_w"], bd, pdn, ap, ps, dt.loc[:, sample+"_wpc"], 1.5, 2)/epsilon_0
    FitsP.at["LindeP_fitted", "RMSE_"+sample] = min(LindeP_RMSE)
    FitsP.at["LindeP_fitted", "R2_"+sample]   = 1 - (np.sum((LindeP_fitted-dt.loc[:, sample+"_p"])**2)/etot)
    print("The optimum m is:", m_fitted)
    ds.at[s, "m_linde"] = m_fitted

    
    """ Dobson's model"""
    for i in range(len(bwp_)):
        
        Dobson_RMSE.append(np.sqrt(np.mean((ppm.dobson85(dt.loc[:, sample+"_w"], bw, bd, pdn, ap, ps, dt.loc[:, sample+"_wpc"], bwp_[i])/epsilon_0 -dt.loc[:, sample+"_p"])**2)))

    bwp_fitted = bwp_[Dobson_RMSE.index(min(Dobson_RMSE))]
    
    Dobson_fitted = ppm.dobson85(dt.loc[:, sample+"_w"], bw, bd, pdn, ap, ps, dt.loc[:, sample+"_wpc"], bwp_fitted)/epsilon_0
    FitsP.at["Dobson_fitted", "RMSE_"+sample] = min(Dobson_RMSE)
    FitsP.at["Dobson_fitted", "R2_"+sample] =   1 - (np.sum((Dobson_fitted-dt.loc[:, sample+"_p"])**2)/etot)
    print("The optimum bwp is:", bwp_fitted)
    ds.at[s, "bwp_dobson"] = bwp_fitted
    
    
    ####################################### Non-fitting models evaluation ################################
    
        
    Hydra_pred, Topp_pred, Jacandschj1993B_pred, Nadler_pred, Logsdonperm_pred, Steelman_pred, Jacandschj1993A_pred = [], [], [], [], [], [], []
    crim_pred, Peplinski95_pred, Malicki_pred= [], [], [], 
    FengSen_Mod_pred, Roth90_Mod_pred, LindeP_Mod_pred, Sen81_Mod_pred, Dobson_mod_pred, EndresRedman_mod_pred, Wund13_mod_pred = [], [], [], [], [], [], []
    
    EndresRedman = ppm.endres_redman(  dt.loc[:, sample+"_w"], bd, pdn, ap, ps, dt.loc[:, sample+"_wpc"], 0.3)/epsilon_0
    FengSen = ppm.feng_sen85(       dt.loc[:, sample+"_w"], bd, pdn, ap, ps, dt.loc[:, sample+"_wpc"],  0.3)/epsilon_0
    Wund13 = ppm.wund_perm(dt.loc[:, sample+"_w"], 4, 0.05, dt.loc[:, sample+"_wpc"],  0.01)/epsilon_0
    Sen81 = ppm.sen1981(           dt.loc[:, sample+"_w"], bd, pdn, ap, ps, dt.loc[:, sample+"_wpc"],  0.3)/epsilon_0
    
    FitsP.at["FengSen", "R2_"+sample] =      1 - (np.sum((FengSen-dt.loc[:, sample+"_p"])**2)/etot)
    FitsP.at["LindeP", "R2_"+sample] =       1 - (np.sum((LindeP - dt.loc[:, sample+"_p"])**2)/etot)
    FitsP.at["Sen81", "R2_"+sample] =        1 - (np.sum((Sen81 - dt.loc[:, sample+"_p"])**2)/etot)
    #FitsP.at["Dobson", "R2_"+sample] =       1 - (np.sum((LindeP_IOP-dt.loc[:, sample+"_p"])**2)/etot)
    FitsP.at["EndresRedman", "R2_"+sample] = 1 - (np.sum((EndresRedman-dt.loc[:, sample+"_p"])**2)/etot)
    FitsP.at["Wund13", "R2_"+sample] =       1 - (np.sum((Wund13-dt.loc[:, sample+"_p"])**2)/etot)
    
    FitsP.at["FengSen", "RMSE_"+sample] =      np.sqrt(np.mean((FengSen - dt.loc[:, sample+"_p"])**2))
    FitsP.at["LindeP", "RMSE_"+sample] =       np.sqrt(np.mean((LindeP - dt.loc[:, sample+"_p"])**2))
    FitsP.at["Sen81", "RMSE_"+sample] =        np.sqrt(np.mean((Sen81 - dt.loc[:, sample+"_p"])**2))
    #FitsP.at["Dobson", "RMSE_"+sample] =       np.sqrt(np.mean((Dobson_mod_pred-dt.loc[:, sample+"_p"])**2))
    FitsP.at["EndresRedman", "RMSE_"+sample] = np.sqrt(np.mean((EndresRedman - dt.loc[:, sample+"_p"])**2))
    FitsP.at["Wund13", "RMSE_"+sample] =       np.sqrt(np.mean((Wund13 - dt.loc[:, sample+"_p"])**2))

    Hydra_pred = ppm.hydraprobe(      dt.loc[:, sample+"_w"]  )/epsilon_0
    Topp_pred =  ppm.topp1980(dt.loc[:, sample+"_w"].all())/epsilon_0
    Jacandschj1993B_pred =  ppm.jacandschj1993B(      dt.loc[:, sample+"_w"].any(), bd, cc, org  )/epsilon_0
    Nadler_pred = ppm.nadler1991(      dt.loc[:, sample+"_w"].any()  )/epsilon_0
    #Logsdonperm_pred = ppm.logsdonperm(      dt.loc[:, sample+"_w"].any()  )/epsilon_0
    Steelman_pred = ppm.steelman(      dt.loc[:, sample+"_w"].any()  )/epsilon_0
    Jacandschj1993A_pred = ppm.jacandschj1993A(      dt.loc[:, sample+"_w"].any()  )/epsilon_0
    
    FitsP.at["Hydra", "RMSE_"+sample]  =          np.sqrt(np.mean((Hydra_pred - dt.loc[:, sample+"_p"])**2))
    FitsP.at["Topp",  "RMSE_"+sample] =           np.sqrt(np.mean(( Topp_pred - dt.loc[:, sample+"_p"])**2))
    FitsP.at["Jacandschj1993B", "RMSE_"+sample] = np.sqrt(np.mean(( Jacandschj1993B_pred - dt.loc[:, sample+"_p"])**2))
    FitsP.at["Nadler", "RMSE_"+sample] =          np.sqrt(np.mean(( Nadler_pred - dt.loc[:, sample+"_p"])**2))
    #FitsP.at["Logsdonperm", "RMSE_"+sample] =     np.sqrt(np.mean(( Logsdonperm_pred - dt.loc[:, sample+"_p"])**2))
    FitsP.at["Steelman", "RMSE_"+sample] =        np.sqrt(np.mean(( Steelman_pred - dt.loc[:, sample+"_p"])**2))
    FitsP.at["Jacandschj1993A", "RMSE_"+sample] = np.sqrt(np.mean(( Jacandschj1993A_pred - dt.loc[:, sample+"_p"])**2))
    
    FitsP.at["Hydra", "R2_"+sample]  =          1 - (np.sum((Hydra_pred-dt.loc[:, sample+"_p"])**2)/etot)                                               
    FitsP.at["Topp",  "R2_"+sample] =           1 - (np.sum((Topp_pred-dt.loc[:, sample+"_p"])**2)/etot)
    FitsP.at["Jacandschj1993B", "R2_"+sample] = 1 - (np.sum((Jacandschj1993B_pred-dt.loc[:, sample+"_p"])**2)/etot)
    FitsP.at["Nadler", "R2_"+sample] =          1 - (np.sum((Nadler_pred-dt.loc[:, sample+"_p"])**2)/etot)
    #FitsP.at["Logsdonperm", "R2_"+sample] =     1 - (np.sum((LindeP_pred-dt.loc[:, sample+"_p"])**2)/etot)
    FitsP.at["Steelman", "R2_"+sample] =        1 - (np.sum((Steelman_pred-dt.loc[:, sample+"_p"])**2)/etot)
    FitsP.at["Jacandschj1993A", "R2_"+sample] = 1 - (np.sum((Jacandschj1993A_pred-dt.loc[:, sample+"_p"])**2)/etot)

    crim_pred = ppm.crim(            dt.loc[:, sample+"_w"], bd, pdn,            ap, ps, dt.loc[:, sample+"_wpc"])/epsilon_0
    Peplinski95_pred = ppm.peplinski(       dt.loc[:, sample+"_w"], bd, pdn, cc, sc, ps, dt.loc[:, sample+"_wpc"], p_inf)/epsilon_0
    Malicki_pred = ppm.malicki(         dt.loc[:, sample+"_w"], bd                                )/epsilon_0
    
    FitsP.at["crim", "RMSE_"+sample] =            np.sqrt(np.mean((crim_pred - dt.loc[:, sample+"_p"])**2))
    FitsP.at["Peplinski95", "RMSE_"+sample] =     np.sqrt(np.mean((Peplinski95_pred - dt.loc[:, sample+"_p"])**2))
    FitsP.at["Malicki", "RMSE_"+sample] =         np.sqrt(np.mean((Malicki_pred - dt.loc[:, sample+"_p"])**2))
    
    FitsP.at["crim", "R2_"+sample] =            1 - (np.sum((crim_pred-dt.loc[:, sample+"_p"])**2)/etot)
    FitsP.at["Peplinski95", "R2_"+sample] =     1 - (np.sum((Peplinski95_pred-dt.loc[:, sample+"_p"])**2)/etot)
    FitsP.at["Malicki", "R2_"+sample] =         1 - (np.sum((Malicki_pred-dt.loc[:, sample+"_p"])**2)/etot)

    Roth90_IOP = ppm.roth1990_IOP(dt.loc[:, sample+"_w"], bd, pdn,  ap, ps, dt.loc[:, sample+"_wpc"], CEC)/epsilon_0
    LindeP_IOP = ppm.lindep_IOP(dt.loc[:, sample+"_w"], bd, pdn,   ap, ps, dt.loc[:, sample+"_wpc"], CEC)/epsilon_0
    Sen81_IOP = ppm.sen81_IOP(dt.loc[:, sample+"_w"], bd, pdn,    ap, ps, dt.loc[:, sample+"_wpc"], CEC)/epsilon_0
    #Dobson_IOP 
    endres_redman_IOP = ppm.endres_redman_IOP(dt.loc[:, sample+"_w"], bd, pdn, ap, ps, dt.loc[:, sample+"_wpc"], CEC)/epsilon_0
    feng_sen85_IOP = ppm.feng_sen85_IOP(dt.loc[:, sample+"_w"], bd, pdn, ap, ps, dt.loc[:, sample+"_wpc"], CEC)/epsilon_0
    wund_perm_IOP = ppm.wund_perm_IOP(dt.loc[:, sample+"_w"], 4, 0.05, dt.loc[:, sample+"_wpc"], CEC)/epsilon_0
    
    FitsP.at["feng_sen85_IOP", "RMSE_"+sample] =      np.sqrt(np.mean((feng_sen85_IOP - dt.loc[:, sample+"_p"])**2))
    FitsP.at["Roth90_IOP", "RMSE_"+sample] =       np.sqrt(np.mean((Roth90_IOP - dt.loc[:, sample+"_p"])**2))
    FitsP.at["LindeP_IOP", "RMSE_"+sample] =       np.sqrt(np.mean((LindeP_IOP - dt.loc[:, sample+"_p"])**2))
    FitsP.at["Sen81_IOP", "RMSE_"+sample] =        np.sqrt(np.mean((Sen81_IOP - dt.loc[:, sample+"_p"])**2))
    #FitsP.at["Dobson_mod", "RMSE_"+sample] =       np.sqrt(np.mean((Dobson_mod_pred-dt.loc[:, sample+"_p"])**2))
    FitsP.at["endres_redman_IOP", "RMSE_"+sample] = np.sqrt(np.mean((endres_redman_IOP - dt.loc[:, sample+"_p"])**2))
    FitsP.at["wund_perm_IOP", "RMSE_"+sample] =       np.sqrt(np.mean((wund_perm_IOP - dt.loc[:, sample+"_p"])**2))

    FitsP.at["feng_sen85_IOP", "R2_"+sample] =      1 - (np.sum((feng_sen85_IOP-dt.loc[:, sample+"_p"])**2)/etot)
    FitsP.at["Roth90_IOP", "R2_"+sample] =       1 - (np.sum((Roth90_IOP - dt.loc[:, sample+"_p"])**2)/etot)
    FitsP.at["LindeP_IOP", "R2_"+sample] =       1 - (np.sum((LindeP_IOP - dt.loc[:, sample+"_p"])**2)/etot)
    FitsP.at["Sen81_IOP", "R2_"+sample] =        1 - (np.sum((Sen81_IOP  - dt.loc[:, sample+"_p"])**2)/etot)
    #FitsP.at["Dobson_IOP", "R2_"+sample] =       1 - (np.sum((LindeP_IOP-dt.loc[:, sample+"_p"])**2)/etot)
    FitsP.at["endres_redman_IOP", "R2_"+sample] = 1 - (np.sum((endres_redman_IOP-dt.loc[:, sample+"_p"])**2)/etot)
    FitsP.at["wund_perm_IOP", "R2_"+sample] =       1 - (np.sum((wund_perm_IOP-dt.loc[:, sample+"_p"])**2)/etot)
    
```
